## First Case Study : Miles Per Gallon Drivers are concerned with the mileage obtained by their automobiles. One driver has kept track of several tankfuls of gasolineby recording miles driven and gallons used for each tankful. Develop a sentinel-controlled-repetition script that prompts theuser to input the miles driven and gallons used for each tankful.
 The script should calculate and display the miles per gallon obtained for each tankful. After processing all input information, thescript should calculate and display the combined miles per gallon obtained for all tankfuls (that is, total miles driven divided bytotal gallons used).
 Requirment Output of The Program :



```python
total_gallons = 0
total_miles = 0

while True:
    gallons = float(input("Enter the gallons used (-1 to end): "))
    if gallons == -1:
        break
    
    miles = float(input("Enter the miles driven: "))
    mpg = miles / gallons
    total_gallons += gallons
    total_miles += miles
    
    print(f"The miles per gallon is {mpg}")

if total_gallons > 0:
    total_mpg = total_miles / total_gallons
    print(f"The overall average miles/gallon was {total_mpg}")
else:
    print("No data to calculate overall average miles/gallon.")
```

## Second Case Study : A Game of Chance - Crap You roll two six-sided dice, each with faces containing one, two, three, four, five and six spots, respectively. When the dice come to rest, the sum of the spots on the two upward faces is calculated. If the sum is 7 or 11 on the first roll, you win. If the sum is 2, 3 or 12 on the first roll (called craps), you lose (i.e., the house wins). If the sum is 4, 5, 6, 8, 9 or 10 on the first roll, that sum becomes your point. To win, you must continue rolling the dice until you make your point (i.e., roll that same point value). You lose by rolling a 7 before making your point.ng your point.

```python
import random

def roll_dice(n_sides=6):
    return random.randint(1, n_sides)

def craps():
    def print_roll(dice_1, dice_2):
        print(f"Roll: {dice_1} + {dice_2} = {dice_1 + dice_2}")

    dice_1, dice_2 = roll_dice(), roll_dice()
    print_roll(dice_1, dice_2)
    sum_roll = dice_1 + dice_2

    if sum_roll in [7, 11]:
        print("You win!")
        return
    elif sum_roll in [2, 3, 12]:
        print("You lose!")
        return

    point = sum_roll
    print(f"Point is {point}")

    while True:
        input("Press Enter to roll the dice")
        dice_1, dice_2 = roll_dice(), roll_dice()
        print_roll(dice_1, dice_2)
        sum_roll = dice_1 + dice_2
        
        if sum_roll == point:
            print("You win!")
            return
        elif sum_roll == 7:
            print("You lose!")
            return

craps()
```

## Third Case Study : COVID 19 Infection Statistic (COVID-19 Infection Statistics) During the first 20 days of the COVID-19 pandemic, the number of newly infected patients per day in a country were recorded. Place the following numbers in a list: 174, 335, 278, 214, 422, 513, 737, 672, 489, 412, 1301, 1105, 1123, 1376, 1502, 894, 665, 1704, 1656, 1342 Use the built-in functions in the statistics module to display the following statistics concerning the infection rate: minimum, maximum, range, mean, median, variance, and standard deviation.d deviation.

```python
import statistics

n_infected_patients = [174, 335, 278, 214, 422, 513, 737, 672, 489, 412, 1301, 1105, 1123, 1376, 1502, 894, 665, 1704, 1656, 1342]

min_value = min(n_infected_patients)
max_value = max(n_infected_patients)
range_value = max_value - min_value
mean_value = statistics.mean(n_infected_patients)
median_value = statistics.median(n_infected_patients)
variance_value = statistics.variance(n_infected_patients)
std_dev_value = statistics.stdev(n_infected_patients)

print(f"The minimum number of patients infected in a day was {min_value}")
print(f"The maximum number of patients infected in a day was {max_value}")
print(f"The range of patients infected in a day was {range_value}")
print(f"The mean number of patients infected in a day was {mean_value}")
print(f"The median number of patients infected in a day was {median_value}")
print(f"The variance in the number of patients infected in a day was {variance_value}")
print(f"The standard deviation in the number of patients infected in a day was {std_dev_value}")
```
